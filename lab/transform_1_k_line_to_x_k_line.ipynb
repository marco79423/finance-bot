```python
import pandas as pd


df = pd.read_csv('../data/TXF1.csv')
df = df.rename(columns={
    'Date': 'date',
    'Time': 'time',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'TotalVolume' : 'volume',
})

df['timestamp'] = pd.to_datetime(df['date'] + ' ' + df['time'])
df.drop(['date', 'time'], axis=1, inplace=True)
df.set_index('timestamp', inplace=True)

df
```

```python
df = df.resample('4T').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
})
df
```
